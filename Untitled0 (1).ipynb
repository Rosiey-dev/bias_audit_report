```python
# Install AIF360 (run only once)
!pip install aif360 --quiet

# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

from aif360.datasets import BinaryLabelDataset
from aif360.metrics import BinaryLabelDatasetMetric, ClassificationMetric
from aif360.algorithms.preprocessing import Reweighing

# Load dataset from UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    "age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
    "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
    "hours-per-week", "native-country", "income"
]
df = pd.read_csv(url, names=column_names, skipinitialspace=True)
df.replace("?", np.nan, inplace=True)
df.dropna(inplace=True)

# Encode categorical columns
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = LabelEncoder().fit_transform(df[col])

label_name = "income"
protected_attr = "sex"  # 1=male (privileged), 0=female (unprivileged)

# Split features and labels
features = df.drop(label_name, axis=1)
labels = df[label_name]

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.3, random_state=42, stratify=labels)

# Scale
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initial model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Initial Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Fairness metrics BEFORE mitigation
dataset_all = pd.concat([features, labels], axis=1)
aif_data = BinaryLabelDataset(df=dataset_all, label_names=[label_name], protected_attribute_names=[protected_attr])

metric = BinaryLabelDatasetMetric(
    aif_data,
    privileged_groups=[{protected_attr: 1}],
    unprivileged_groups=[{protected_attr: 0}]
)

print("Statistical Parity Difference (Before):", metric.statistical_parity_difference())
print("Disparate Impact (Before):", metric.disparate_impact())

# Prepare for mitigation
X_cols = features.columns
train_df = pd.concat([pd.DataFrame(X_train, columns=X_cols), pd.Series(y_train.reset_index(drop=True), name=label_name)], axis=1)
test_df = pd.concat([pd.DataFrame(X_test, columns=X_cols), pd.Series(y_test.reset_index(drop=True), name=label_name)], axis=1)

train_dataset = BinaryLabelDataset(df=train_df, label_names=[label_name], protected_attribute_names=[protected_attr])
test_dataset = BinaryLabelDataset(df=test_df, label_names=[label_name], protected_attribute_names=[protected_attr])

# Reweighing
rw = Reweighing(unprivileged_groups=[{protected_attr: 0}], privileged_groups=[{protected_attr: 1}])
train_rw = rw.fit_transform(train_dataset)

# Train with weights
model_rw = LogisticRegression(max_iter=1000)
model_rw.fit(X_train, y_train, sample_weight=train_rw.instance_weights)
y_pred_rw = model_rw.predict(X_test)

# Evaluate after mitigation
test_pred_rw = test_dataset.copy()
test_pred_rw.labels = y_pred_rw.reshape(-1, 1).astype(int)

metric_rw = ClassificationMetric(
    test_dataset, test_pred_rw,
    privileged_groups=[{protected_attr: 1}],
    unprivileged_groups=[{protected_attr: 0}]
)

print("\n== AFTER Reweighing ==")
print("Accuracy:", accuracy_score(y_test, y_pred_rw))
print("Statistical Parity Difference:", metric_rw.statistical_parity_difference())
print("Disparate Impact:", metric_rw.disparate_impact())

# Fairness comparison bar chart
metrics_names = ['Accuracy', 'Statistical Parity Diff', 'Disparate Impact']
before_values = [accuracy_score(y_test, y_pred), metric.statistical_parity_difference(), metric.disparate_impact()]
after_values = [accuracy_score(y_test, y_pred_rw), metric_rw.statistical_parity_difference(), metric_rw.disparate_impact()]

x = range(len(metrics_names))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar([i - width/2 for i in x], before_values, width, label='Before Mitigation', color='salmon')
ax.bar([i + width/2 for i in x], after_values, width, label='After Mitigation', color='skyblue')

ax.set_xticks(x)
ax.set_xticklabels(metrics_names)
ax.set_title('Fairness Metrics Before and After Mitigation')
ax.legend()

for i in x:
    ax.text(i - width/2, before_values[i] + 0.01, f"{before_values[i]:.3f}", ha='center')
    ax.text(i + width/2, after_values[i] + 0.01, f"{after_values[i]:.3f}", ha='center')

plt.ylim([0, 1.1])
plt.show()

# Outcome distribution plotting
def plot_outcome_distribution(df, title):
    grouped = df.groupby(['sex', 'income']).size().unstack(fill_value=0)

    labels = ['Female (0)', 'Male (1)']
    favorable = grouped[1] if 1 in grouped.columns else [0, 0]
    unfavorable = grouped[0] if 0 in grouped.columns else [0, 0]

    x = np.arange(len(labels))
    width = 0.35

    fig, ax = plt.subplots()
    ax.bar(x - width/2, favorable, width, label='Favorable (1)', color='blue')
    ax.bar(x + width/2, unfavorable, width, label='Unfavorable (0)', color='red')

    ax.set_ylabel('Count')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    plt.ylim(0, max(favorable.max(), unfavorable.max()) * 1.2)
    plt.show()

# Get BEFORE mitigation predictions
df_before = test_dataset.convert_to_dataframe()[0]

# Get AFTER mitigation predictions
df_after = test_pred_rw.convert_to_dataframe()[0]

# Plot
plot_outcome_distribution(df_before, "Outcome Distribution Before Mitigation")
plot_outcome_distribution(df_after, "Outcome Distribution After Mitigation")
```
